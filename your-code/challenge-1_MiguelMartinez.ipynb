#### How many times does each letter in `letters` appear in `tup3`?

Print out the number of occurrence of each letter.

````python
count_letters = {}

letters = ["a", "b", "c", "d", "e"]

for element in letters:
    if element in count_letters.keys():
        count_letters[element] += 1
    else:
        count_letters[element] = 1
        
print(count_letters)
````
