```python
pwd = getpass.getpass()
cnx = mysql.connector.connect(user = 'root',
                              password = pwd,
                              host = 'localhost',
                              database = 'sakila')
```

```python
#we need to define the object we will use to interact with database

cursor = cnx.cursor()
cursor
```

```python
#now we want to create a query
# we will write code in SQL syntax

query = ("SELECT * FROM sakila.actor;")
```

```python
#how do we exacute this query?

cursor.execute(query)

print(cursor)
```

```python
#now we need to bring the results back to python

results = cursor.fetchall()
results
```

```python
query = (""" CREATE DATABASE IF NOT EXISTS newdb""")
cursor.execute(query)
```

```python
##Create a table

query = ("""CREATE TABLE IF NOT EXISTS
newdb.country(
country_id INT NOT NULL PRIMARY KEY,
country_name VARCHAR(52) NOT NULL
); """)

cursor.execute(query)
```

```python
## Create a table, that will connect with the table country

query = ("""CREATE TABLE IF NOT EXISTS
newdb.demo(
    student_id INT PRIMARY KEY,
    student_name VARCHAR(52),
    email VARCHAR(52) DEFAULT "johndoe@ironhack.com",
    coolness INT NOT NULL,
    country_id INT NOT NULL,
    FOREIGN KEY (country_id) REFERENCES country(country_id));""")

cursor.execute(query)
```

```python
query = (""" CREATE TABLE IF NOT EXISTS
newdb.fullactor(
    actor_id INT PRIMARY KEY,
    fullname VARCHAR(52),
    coolness INT NOT NULL DEFAULT 5); """)

cursor.execute(query)
```
