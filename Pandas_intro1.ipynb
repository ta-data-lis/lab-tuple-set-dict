# Groupby

The groupby method allows you to group rows of data together and call aggregate functions. It would be the equivalent compared to the Pivot tables.

```python
operations = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Sam','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
df = pd.DataFrame(operations)
df
```

```python
df.groupby("Company").mean()
```

```python
df.groupby("Company").std()
```

```python
df.groupby("Company").min()
```

```python
df.groupby("Company").count()
```

```python
df.groupby("Company").nunique()
```

```python
df.groupby("Company").describe()
```

## Dealing with nulls/nans;
Further information about nan nature in Python: https://towardsdatascience.com/navigating-the-hell-of-nans-in-python-71b12558895b

```python
operations = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Sam','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,np.nan]}
df = pd.DataFrame(operations)
df
```

```python
df.groupby("Company").mean()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.fillna(value='FILL VALUE')
```
